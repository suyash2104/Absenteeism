```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
raw_csv_data.head()
```

```python
raw_csv_data.describe(include='all')
```

```python
df = raw_csv_data.copy()
df.info()
```

```python
df = df.drop(['ID'], axis=1)
```

```python
len(df['Reason for Absence'].unique())
```

```python
reason_column = pd.get_dummies(df['Reason for Absence'], drop_first=True)
```

```python
reason_column
```

```python
reason_type_1 = reason_column.loc[:,1:14].max(axis=1)
reason_type_2 = reason_column.loc[:,15:17].max(axis=1)
reason_type_3 = reason_column.loc[:,18:21].max(axis=1)
reason_type_4 = reason_column.loc[:,22:].max(axis=1)
```

```python
reason_type_1
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df = df.drop(['Reason for Absence'], axis=1)
```

```python
df
```

```python
df.columns.values
```

```python
column_values = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
df.columns = column_values
```

```python
df.head()
```

```python
column_values = [ 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_values]
df
```

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
len(df_reason_mod)
```

```python
list_months = []
for i in range(700):
    list_months.append(df_reason_mod['Date'][i].month)
list_months
```

```python
df_reason_mod['Months values'] = list_months
df_reason_mod
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
```

```python
df_reason_mod.columns.values
```

```python
column_values = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Months values',
       'Day of the week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Months values',
       'Day of the week']
df_reason_mod = df_reason_mod[column_values]
```

```python
df_reason_mod
```

```python
df_reason_date_mod = df_reason_mod.copy()
```

```python
df_reason_date_mod['Education'] = df_reason_date_mod["Education"].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_reason_date_mod['Education'].unique()
```

```python
df_preprocessed = df_reason_mod.copy()
```

```python
df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index=False)
```
